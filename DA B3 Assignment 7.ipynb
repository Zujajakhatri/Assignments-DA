### Importing libraries and reading the file

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("Salaries.csv")
df
```

1. Which employee has the highest TotalPayBenefits?

```python
highest_paid_employee = df.loc[df['TotalPayBenefits'].idxmax()]
highest_paid_employee
```

2. Drop ID, Notes, Agency and Status Columns

```python
df=df.drop(["Id", "Notes", "Agency","Status"],axis=1)
df
```

3. Total Number of Job Titles Contain POLICE

```python
police_jobs = df['JobTitle'].str.contains('POLICE', case=False, na=False).sum()
print ("Total number of job titles contains POLICE:", police_jobs)
```

4. Top 10 most common job titles?

```python
common_job_titles=df['JobTitle'].value_counts().head(10)
common_job_titles
```

5. What are the top 10 average TotalPay by JobTitle?

```python
df['TotalPay'] = pd.to_numeric(df['TotalPay'], errors='coerce')

# Group by JobTitle and calculate the average TotalPay
avg_total_pay_by_job = df.groupby('JobTitle')['TotalPay'].mean()

# Get the top 10 job titles with highest average TotalPay
top_10_avg_pay = avg_total_pay_by_job.sort_values(ascending=False).head(10)
top_10_avg_pay
```
